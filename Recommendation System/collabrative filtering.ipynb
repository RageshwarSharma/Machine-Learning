```python
# to ensure stastical significance user with less then 200 ratings and books with less then 100 ratings will be excluded
```

```python
count1=ratings['userID'].value_counts()
ratings=ratings[ratings['userID'].isin(count1[count1>=200].index)]
counts=ratings['bookRating'].value_counts()
ratings=ratings[ratings['bookRating'].isin(counts[counts>=100].index)]
```

```python
ratings.head()
```

```python
# colabrative filtering using k nearest neighbour
## it will help to find the cluster of similar user based on common book ratings
```

```python
# merge ratings and columns table
combine_book_rating=pd.merge(ratings,books,on='ISBN')
```

```python
columns=['bookAuthor','yearofPublication','publisher','imageUrlS','imageUrlM','imageUrlL']
combine_book_rating=combine_book_rating.drop(columns,axis=1)
```

```python
combine_book_rating
```

```python
## combining the book title and making new column for total count
```

```python
combine_book_rating=combine_book_rating.dropna(axis=0,subset=['bookTitle'])

book_rating_count=(combine_book_rating.groupby(by=['bookTitle'])['bookRating'].count().reset_index().
                  rename(columns={'bookRating':'totalratingcount'})
                  [['bookTitle','totalratingcount']])
book_rating_count.head()
```

```python
# combining rating data with the total rating count data, this gives us which book are popular and filter out less popular book
```

```python
rating_with_totalRatingcount=combine_book_rating.merge(book_rating_count,left_on='bookTitle'
                                                       ,right_on='bookTitle',how='left')
rating_with_totalRatingcount.head()
```

```python
# setting up the threshold value
# if book has rating greater than 50 that book will be count

popularity_threshold=50
rating_popular_book=rating_with_totalRatingcount.query('totalratingcount>=@popularity_threshold')
rating_popular_book.head()
```

```python
rating_popular_book.shape
```

```python
# filter to use in US and Canada Only
```

```python
combined=rating_popular_book.merge(user,left_on='userID',right_on='userID',how='left')
us_canada_user_rating=combined[combined['Location'].str.contains('usa|canada')]
us_canada_user_rating=us_canada_user_rating.drop('Age',axis=1)
us_canada_user_rating.head()
```

```python
# Applying Cosine Similarity
```

```python
# we convert the table in 2d and fill the missing value with zero
```

```python
# for sparse matrix
from scipy.sparse import csr_matrix
us_canada_user_rating=us_canada_user_rating.drop_duplicates(['userID','bookTitle'])
us_canada_user_rating_pivot=us_canada_user_rating.pivot(index='bookTitle',columns='userID',values='bookRating').fillna(0)
us_canada_user_rating_matrix=csr_matrix(us_canada_user_rating_pivot.values)
```

```python
us_canada_user_rating_pivot.head()
```

```python
# Applying K nearset Neighbour
```

```python
from sklearn.neighbors import NearestNeighbors
model_knn=NearestNeighbors(metric='cosine',algorithm='brute')
model_knn.fit(us_canada_user_rating_matrix)
```

```python
us_canada_user_rating_pivot
```

```python
# Selecting the random value
# Here we selecting the n_neighbours 6 so we will get 6 recommendation
```

```python
query_index=np.random.choice(us_canada_user_rating_pivot.shape[0])
print(query_index)
distances,indices=model_knn.kneighbors(us_canada_user_rating_pivot.iloc[query_index,:].values.reshape(1,-1),n_neighbors=6)
```

```python
us_canada_user_rating_pivot.index[query_index]
```

```python
# recommendation we get from Mortal Prey are given
```

```python
for i in range(0,len(distances.flatten())):
    if i==0:
        print('Recommendation for {0}:\n'.format(us_canada_user_rating_pivot.index[query_index]))
    else:
        print('{0} : {1} with distance of {2}'.format(i, us_canada_user_rating_pivot.index[indices.flatten()[i]],distances.flatten()[i]))
```
