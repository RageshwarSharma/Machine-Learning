### Importing Required Libraries

```python
import pandas as pd
import tweepy
from tweepy import OAuthHandler
from tweepy import StreamListener
from tweepy import Stream
import csv
```

## Specifing keys and tokens 

### Getting key and token from twitter developer account

```python
consumer_key='__________________________'
consumer_secret = '______________________________________'
access_token = '______________________________________________'
access_token_secret = '___________________________________________'
```

## Getting authentication from twitter using keys and token

```python
# authorization key
auth=OAuthHandler(consumer_key,consumer_secret)
```

```python
# access keys
auth.set_access_token(access_token, access_token_secret)
```

### Declaring class for handling stream data

```python
class tweet(StreamListener):
    
    def on_status(self,status):
        if (status.retweeted):
            return True
        
        User_id=status.id_str
        User_name=status.user.screen_name
        User_description=status.user.description
        Tweet=status.text
        User_loc=status.user.location
        
        followers=status.user.followers_count
        
        with open("streamed_tweets.csv",mode='a',newline='\n',encoding='utf-8') as f:
            writer=csv.writer(f)
            writer.writerow([User_id,User_name,User_description,Tweet,User_loc,followers])
            
    def on_error(self,status):
        print(status)
Listner = tweet()
```

```python
stream = Stream(auth=auth,listener=Listner)
```

## Live Stream data for Corona-Virus using different hashtags

```python
stream.filter(track=['coronoavirus','CoronavirusOutbreak','coronavirusPH','WuhanVirus','CoronaVirus'
                    ,'ChinaVirus','ChinaCoronaVirus'])
```
